```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path_to_folder = 'gdrive/MyDrive/MathWorks/'
employment_path = path_to_folder + 'employment_data/'
```

```python
employment_data = [pd.read_csv(employment_path + 'washington.csv'), 
                   pd.read_csv(employment_path + 'pennsylvania.csv'),
                   pd.read_csv(employment_path + 'nebraska.csv'),
                   pd.read_csv(employment_path + 'england.csv'),
                   pd.read_csv(employment_path + 'wales.csv')]
```

```python
sum = [0 for i in range(10)]
data = [[0 for j in range(10)] for i in range(5)]
for city in range(len(employment_data)):
  for value in range(len(employment_data[city]['2021'])):
    sum[value] += employment_data[city]['2021'][value]
    data[city][value] = employment_data[city]['2021'][value]
percent_data = [[0 for j in range(10)] for i in range(5)]
for city in range(len(data)):
  for i in range(len(data[city])):
    percent_data[city][i] = data[city][i]/sum[i]
percent_dict = {
    'occupation': employment_data[0]['Unnamed: 0'],
    'washington': percent_data[0],
    'pennsylvania': percent_data[1],
    'nebraska': percent_data[2],
    'england': percent_data[3],
    'wales': percent_data[4]
}
df = pd.DataFrame(percent_dict)
df
```

```python
weighted_remote_data = {
    'Mining, logging, construction': 0.19,
    'Manufacturing': 0.01,
    'Trade, transportation, and utilities': 0.03,
    'Information': 0.72,
    'Financial activities': 0.76,
    'Professional and business services': 0.8,
    'Education and health services': 0.83,
    'Leisure and hospitality': 0.71,
    'Other services': 0.33,
    'Government': 0.35
}
```

```python
demographics_path = path_to_folder + 'demographics_data/'
for location in ['washington', 'wales', 'pennsylvania', 'nebraska', 'england']:
  max = sorted(zip(df['occupation'], df[location]), reverse=True)[:3]
  max = [i[0] for i in max]

  demographics = pd.read_csv(demographics_path + location + '.csv')
  l = list(demographics[demographics.columns[1]][5:10])
  l = [float(i[:-1])/100 for i in l]
  sum = 0
  for n in range(len(l)):
    sum += (n+1)*l[n]

  print('{}: {}, {}({}), {}({}), {}({})'.format(location, sum, max[0], weighted_remote_data[max[0]], max[1], weighted_remote_data[max[1]], max[2], weighted_remote_data[max[2]]))
```
